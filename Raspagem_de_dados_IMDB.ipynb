3. Utilize a técnica de análise de sentimento nas revisões de modo
a detectar as opiniões contidas nos textos.

```python
from textblob import TextBlob

# Função para realizar análise de sentimento em uma revisão
def analyze_sentiment(review):
    # Criar um objeto TextBlob com o texto da revisão
    blob = TextBlob(review)

    # Calcular a polaridade do sentimento (valor entre -1 e 1)
    polarity = blob.sentiment.polarity

    # Classificar o sentimento como positivo, negativo ou neutro com base na polaridade
    if polarity > 0:
        sentiment = 'Positivo'
    elif polarity < 0:
        sentiment = 'Negativo'
    else:
        sentiment = 'Neutro'

    return sentiment

# Aplicar a função de análise de sentimento em cada revisão e adicionar os resultados a uma nova coluna no DataFrame
df['Sentiment'] = df['Review Text'].apply(analyze_sentiment)

# Exibir as primeiras linhas do DataFrame com a nova coluna de sentimento
print(df.head(20))
```

```python
# Agrupar as revisões por filme e concatená-las em um único texto
grouped_reviews = df.groupby('Movie Title')['Review Text'].apply(lambda x: ' '.join(x)).reset_index()

# Renomear a coluna para 'Combined Review Text'
grouped_reviews.rename(columns={'Review Text': 'Combined Review Text'}, inplace=True)

# Exibir as primeiras linhas do DataFrame resultante
print(grouped_reviews.head())
```

5. Aplique as métricas de similaridade de documentos como:
similaridade do cosseno e distância euclidiana.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances

# Criar um vetorizador TF-IDF
vectorizer = TfidfVectorizer()

# Aplicar o vetorizador aos textos combinados dos filmes
tfidf_matrix = vectorizer.fit_transform(grouped_reviews['Combined Review Text'])

# Calcular a similaridade do cosseno entre os textos
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Calcular a distância euclidiana entre os textos
euclidean_dist = euclidean_distances(tfidf_matrix, tfidf_matrix)

# Exibir a matriz de similaridade do cosseno
print("Matriz de Similaridade do Cosseno:")
print(cosine_sim)

# Exibir a matriz de distância euclidiana
print("\nMatriz de Distância Euclidiana:")
print(euclidean_dist)
```

6. Faça um comando input que aceita uma entrada de texto e,
calcule a similaridade do texto digitado com as revisões
coletadas dos filmes

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Função para calcular a similaridade entre o texto de entrada e as revisões dos filmes
def calculate_similarity(input_text, tfidf_matrix):
    # Converter o texto de entrada em uma representação vetorial usando o mesmo vetorizador TF-IDF
    input_vector = vectorizer.transform([input_text])

    # Calcular a similaridade do cosseno entre o texto de entrada e as revisões dos filmes
    similarity_scores = cosine_similarity(input_vector, tfidf_matrix)

    return similarity_scores

# Entrada do usuário
input_text = input("Digite o texto para calcular a similaridade com as revisões dos filmes: ")

# Calcular a similaridade entre o texto de entrada e as revisões dos filmes
similarity_scores = calculate_similarity(input_text, tfidf_matrix)

# Exibir os resultados de similaridade para cada filme
for i, title in enumerate(grouped_reviews['Movie Title']):
    similarity_score = similarity_scores[0][i]  # Similaridade para o filme atual
    print(f"Similaridade com '{title}': {similarity_score}")
```

7. Ordene as revisões de acordo com a maior similaridade ou menor
distância.

```python
import numpy as np

# Função para ordenar as revisões de acordo com a maior similaridade
def sort_reviews_by_similarity(similarity_scores, titles):
    # Obter os índices das revisões ordenadas por similaridade decrescente
    sorted_indices = np.argsort(similarity_scores)[0][::-1]

    # Ordenar os títulos dos filmes e os scores de similaridade correspondentes
    sorted_titles = [titles[i] for i in sorted_indices]
    sorted_similarity_scores = [similarity_scores[0][i] for i in sorted_indices]

    return sorted_titles, sorted_similarity_scores

# Ordenar as revisões de acordo com a maior similaridade
sorted_titles, sorted_similarity_scores = sort_reviews_by_similarity(similarity_scores, grouped_reviews['Movie Title'])

# Exibir os resultados ordenados por similaridade
for title, similarity_score in zip(sorted_titles, sorted_similarity_scores):
    print(f"Filme: {title} | Similaridade: {similarity_score}")
```

```python
# Função para ordenar as revisões de acordo com a menor distância
def sort_reviews_by_distance(distance_matrix, titles):
    # Obter os índices das revisões ordenadas por distância crescente
    sorted_indices = np.argsort(distance_matrix)[0]

    # Ordenar os títulos dos filmes e as distâncias correspondentes
    sorted_titles = [titles[i] for i in sorted_indices]
    sorted_distances = [distance_matrix[0][i] for i in sorted_indices]

    return sorted_titles, sorted_distances

# Ordenar as revisões de acordo com a menor distância
sorted_titles, sorted_distances = sort_reviews_by_distance(euclidean_dist, grouped_reviews['Movie Title'])

# Exibir os resultados ordenados por distância
for title, distance in zip(sorted_titles, sorted_distances):
    print(f"Filme: {title} | Distância: {distance}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import nltk
from nltk.corpus import stopwords
import pandas as pd

# Carregue sua base de dados, se ainda não estiver carregada
# Supondo que você já tenha uma DataFrame chamada 'DF' com uma coluna 'Review Text'
# Substitua 'dataframe' pelo seu próprio DataFrame, se necessário
# DF = ...

# Inicialize o vetorizador TF-IDF
vectorizer = TfidfVectorizer(stop_words=stopwords.words('portuguese'))

# Transforme os dados de texto em uma matriz TF-IDF
X = vectorizer.fit_transform(df['Review Text'])

# Defina o número de clusters
k = 4

# Inicialize o modelo K-means
kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)

# Ajuste o modelo aos dados
kmeans.fit(X)

# Obtenha os rótulos de cluster para cada amostra
labels = kmeans.labels_

# Adicione os rótulos de cluster ao DataFrame original
df['Cluster'] = labels

# Visualize os clusters resultantes
print(df[['Review Text', 'Cluster']])
```

```python
# Iterar sobre os clusters e imprimir algumas amostras de cada um
for cluster_num in range(k):
    cluster_samples = df[df['Cluster'] == cluster_num].sample(n=5)  # Altere o valor de 'n' conforme necessário
    print(f"Cluster {cluster_num}:\n")
    for index, row in cluster_samples.iterrows():
        print(row['Review Text'])
        print("-" * 50)
    print("\n\n")
```
