```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder #Transformar classes em numeros
from sklearn.feature_extraction.text import CountVectorizer #Cria Vetor com o texto
from sklearn.metrics import confusion_matrix
from keras.models import Sequential # Conjunto de camadas sequenciais
from keras.layers import Dense, Dropout
from google.colab import files
```

```python
# Lê o arquivo CSV
df = pd.read_csv('spam.csv', encoding='ISO-8859-1')  # Use a codificação adequada

# Remove colunas que começam com 'Unnamed'
df = df.loc[:, ~df.columns.str.startswith('Unnamed')]

# Visualiza as primeiras linhas do DataFrame modificado
print(df.head())
```

```python
df.rename(columns={'v1': 'category', 'v2': 'message'}, inplace=True)
```

```python

df.head()
```

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(df['category'])
print(y)
```

Divisão dos dados entre treino e teste

```python
mensagens = df['message'].values
X_train, X_test, y_train, y_test = train_test_split(mensagens, y, test_size=0.3)
print(X_train)
```

```python
vetorizador = CountVectorizer()# Cria vetor de 0 e 1 para cada palavra (OneHot Encoding)
vetorizador.fit(X_train)#Fit cria o modelo
X_train = vetorizador.transform(X_train)
X_test = vetorizador.transform(X_test)
print(X_train.toarray())
```

Visualizar vetor X_train

```python
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
X_train.toarray()[0]
```

```python
print(X_train.shape[1])
```

```python
modelo = Sequential() #adicionar as camadas
#Camada de entrada
# units: neuronios; activation; funcao de ativação; input_dim: qtdade de colunas
modelo.add(Dense(units=10, activation='relu',input_dim=X_train.shape[1]))
#Tecnica para minimizar overfit
#Remove aleatoriamente 1% de sinapses (conexoes)
modelo.add(Dropout(0.1))
#2º camada oculta densa
#nº de neuronios ser referente ao numero de classes
modelo.add(Dense(units=8, activation='relu'))
modelo.add(Dropout(0.1))
#Ultima camada
#Como vai retornar somente 1 valor, units = 1; Muda função de ativação
modelo.add(Dense(units=1, activation='sigmoid'))
```

```python
#Compliar a rede
#Função loss: função de perda
#optmizer: Juste dos pesos. "adam": estocástico
#metrics: Avaliar performance
modelo.compile(loss="mean_squared_error", optimizer="adam", metrics=['accuracy'])
modelo.summary()
```

```python
#Dados de treino
#epochs: serão passados todos os dados de treinos pela rede
#batch_size: A cada qtos registros atualiza-se a rede?
#verbose: retorno do processo iterativo de treinamento
#validation_data: treina e usa dados de validação para melhorar o modelo

modelo.fit(X_train, y_train, epochs=20, batch_size=10, verbose = True, validation_data=(X_test, y_test))
```

Métricas

```python
from keras.src.metrics import accuracy
loss, accuracy = modelo.evaluate(X_test, y_test)
print("loss: ", loss)
print("Acurácia: ", accuracy)
```

```python
nova_previsao = modelo.predict(X_test)
nova_previsao
```

```python
prev = (nova_previsao > 0.5)
prev
```

```python
cm = confusion_matrix(y_test, prev)
cm
```

## Rede Neural 2 - Deep Learning


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder #Transformar classes em numeros
from sklearn.feature_extraction.text import CountVectorizer #Cria Vetor com o texto
from sklearn.metrics import confusion_matrix
from keras.models import Sequential # Conjunto de camadas sequenciais
from keras.layers import Dense, Dropout, Flatten, Embedding
from google.colab import files
```

```python
!pip install keras_preprocessing
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
```

Transformaçao do texto em dados tabulares

```python
labelencoder = LabelEncoder()
y = labelencoder.fit_transform(df['category'])
print(y)
```

```python
mensagens = df['message'].values
X_train, X_test, y_train, y_test = train_test_split(mensagens, y, test_size=0.3)
```

```python
# 1 - Tokenizar o texto
token = Tokenizer(num_words = 1000)#nº max de palavras a serem mantidas
#Fit: transforma
token.fit_on_texts(X_train)
#Criação da matriz densa dos tokens
X_train = token.texts_to_sequences(X_train)
X_test = token.texts_to_sequences(X_test)
```

```python
#Matriz densa dos token
#Transformou palavras em numeros
print(X_train)
```

```python
#Garantir que todos os tokens tenham a mesma quantidade de elementos numericos
#Preenche com zeros depois do final da numeração
# Tamanho maximo: 500 caracteres
X_train = pad_sequences(X_train, padding="post", maxlen=500)
X_test = pad_sequences(X_test, padding="post", maxlen=500)
```

```python
# Dados em forma tabular
print(X_train)
```

```python
print(len(token.word_index))
```

```python
modelo = Sequential()
#Camada Embedding?
#input_dim: tamanho vocabulario
#Output_dim: tamanho do vetor denso
#input_length: comprimento da sequencia
modelo.add(Embedding(input_dim=len(token.word_index), output_dim=50, input_length=500))
modelo.add(Flatten())#Conecta camada Embedding com camada densa

modelo.add(Dense(units=10, activation="relu"))
modelo.add(Dropout(0.1))
modelo.add(Dense(units=1, activation="sigmoid"))
```

```python
modelo.compile(loss="mean_squared_error", optimizer="adam", metrics=["accuracy"])
modelo.summary()
```

```python
modelo.fit(X_train, y_train, epochs=20, batch_size=10,verbose=True, validation_data=(X_test,y_test))
```

```python
from keras.src.metrics import accuracy
loss, accuracy = modelo.evaluate(X_test, y_test)
print("loss: ", loss)
print("Acurácia: ", accuracy)
```

```python
nova_previsao = modelo.predict(X_test)
nova_previsao
```

```python
prev = (nova_previsao > 0.5)
prev
```

```python
cm = confusion_matrix(y_test, prev)
cm
```
